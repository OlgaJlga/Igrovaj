Загружаем данные в датафрейм

```python
import io
import pandas as pd
import numpy as np
import plotly.express as px

df = pd.read_csv(io.BytesIO(uploaded['dataset.csv']))

df
```

```python
df['event_date'] = pd.to_datetime(df['event_date'])
df['install_date'] = pd.to_datetime(df['install_date'])
df.dtypes
```

Для расчета retention rate необходимо знать день активности игрока в каждый конкретный день

```python
df['activity_day'] = (df['event_date'] - df['install_date']) / np.timedelta64 ( 1 , 'D')
df
```

Построим общий retention rate для всех пользователей для дней активности 0-30.

Для того чтобы получить чистые значения уберем из данных инсталлы по несозревшим когортам.

```python
df_ret = df[(df['install_date'] >= '2022-04-01') & (df['install_date'] <= '2022-07-01') & (df['activity_day'] >= 0) & (df['activity_day'] <= 30)][['activity_day', 'appsflyer_id']]
df_ret
```

```python
df_ret_users = df_ret.groupby('activity_day', as_index=False).agg({'appsflyer_id':'nunique'}).rename(columns = {'appsflyer_id':'users'})
df_ret_users
```

Рассчитаем retention rate

```python
installs = df_ret_users[df_ret_users['activity_day']==0]['users'].values[0]

df_ret_users['ret_rate'] = df_ret_users['users']/installs
df_ret_users
```

```python
fig_ret = px.line(df_ret_users, x='activity_day', y='ret_rate')
fig_ret.show()
```

Что можно сказать в целом о краткосрочном / среднесрочном  / длительном удержании в игре?

Посмотрим удержание в игре органики и платных пользователей.

```python
df_ret_channel = df[(df['install_date'] >= '2022-04-01') & (df['install_date'] <= '2022-07-01') & (df['activity_day'] >= 0) & (df['activity_day'] <= 30)][['activity_day', 'is_organic', 'appsflyer_id']]
df_ret_channel = df_ret_channel.groupby(['activity_day', 'is_organic'], as_index = False).agg({'appsflyer_id':'nunique'}).rename(columns = {'appsflyer_id':'users'})
df_ret_channel
```

```python
df_installs_channel = df_ret_channel.groupby('is_organic', as_index=False).agg({'users':'max'}).rename(columns= {'users':'installs'})
df_ret_channel = pd.merge(df_ret_channel, df_installs_channel, on = 'is_organic')
df_ret_channel['ret_rate'] = df_ret_channel['users']/df_ret_channel['installs']

df_ret_channel
```

```python
fig_channel = px.line(df_ret_channel, x='activity_day', y='ret_rate', color='is_organic')

fig_channel.show()
```

Какие выводы можно сделать о retention органики и платных пользователей?

Рассчитаем rolling retention. Для этого нам понадобится дополнительный датафрейм с набором activity day от 0 до 30

```python
df_days = pd.DataFrame({'lifeday_rolling' : range(0,31)})
df_days
```

```python
df_lifetime = df[(df['install_date'] >= '2022-04-01') & (df['install_date'] <= '2022-07-01') & (df['activity_day'] >= 0)].groupby('appsflyer_id', as_index=False).agg({'activity_day':'max'}).rename(columns = {'activity_day':'lifetime'})
df_lifetime
```

Также нам понадобится дополнительный столбец по которому мы сможем сделать merge двух датафреймов

```python
df_days['for_merge'] = 1
df_lifetime['for_merge'] = 1

df_rolling = pd.merge(df_days, df_lifetime, on='for_merge')
df_rolling = df_rolling[df_rolling['lifetime']>=df_rolling['lifeday_rolling']]

df_rolling
```

```python
df_rolling = df_rolling.groupby('lifeday_rolling', as_index = False).agg({'appsflyer_id':'nunique'}).sort_values(by = 'lifeday_rolling').rename(columns = {'appsflyer_id':'users'})
df_rolling['rol_ret'] = round(df_rolling['users'] / df_rolling['users'][0], 2)

df_rolling
```

```python
fig_rolling = px.line(df_rolling, x = 'lifeday_rolling', y = 'rol_ret')
fig_rolling
```

Постройте retention rate первых 30 дней жизни игроков в разрезах: органика/неорганика и определите какой из гео наиболее интересен для каждой когорты. **комментарий от семинариста - я перепутала названия, расчет в файле правильный, его можно использовать

```python
# Фильтрация данных с учетом колонки is_organic и гео
df_ret_geo_channel = df[(df['install_date'] >= '2022-04-01') & (df['install_date'] <= '2022-07-01') & (df['activity_day'] >= 0) & (df['activity_day'] <= 30)][['activity_day', 'appsflyer_id', 'is_organic', 'country_code']]

# Группировка данных по дню активности, каналу (органика или платный) и гео с подсчетом уникальных пользователей
df_ret_geo_channel = df_ret_geo_channel.groupby(['activity_day', 'is_organic', 'country_code'], as_index=False).agg({'appsflyer_id':'nunique'}).rename(columns={'appsflyer_id':'users'})

# Расчет общего числа установок для каждого гео и канала
df_installs_geo_channel = df_ret_geo_channel[df_ret_geo_channel['activity_day'] == 0].drop(columns=['activity_day'])
df_installs_geo_channel = df_installs_geo_channel.rename(columns={'users':'installs'})

# Объединение таблиц и расчет retention rate для каждого дня активности
df_ret_geo_channel = pd.merge(df_ret_geo_channel, df_installs_geo_channel, on=['is_organic', 'country_code'])
df_ret_geo_channel['ret_rate'] = df_ret_geo_channel['users'] / df_ret_geo_channel['installs']

# Фильтрация данных для определения наиболее интересного гео для каждой когорты
# Мы определяем интерес на основе среднего retention rate на протяжении 30 дней
df_ret_avg = df_ret_geo_channel.groupby(['is_organic', 'country_code'], as_index=False).agg({'ret_rate':'mean'}).sort_values(by=['is_organic', 'ret_rate'], ascending=[True, False])

# Определение наиболее интересного гео для органики и неорганики
top_geo_organic = df_ret_avg[df_ret_avg['is_organic'] == True].iloc[0]['country_code']
top_geo_non_organic = df_ret_avg[df_ret_avg['is_organic'] == False].iloc[0]['country_code']

top_geo_organic, top_geo_non_organic
```
