# Вспомним когортный отчет с последнего семинара ⏰

```python
т  тimport io
import pandas as pd

df = pd.read_csv(io.BytesIO(uploaded['dataset.csv']))
df_spend = pd.read_csv(io.BytesIO(uploaded['dataset_cost.csv']))
df_ads = pd.read_csv(io.BytesIO(uploaded['dataset_ad.csv']))

display(df)
display(df_spend)
display(df_ads)
```

```python
df['event_date'] = pd.to_datetime(df['event_date'])
df['install_date'] = pd.to_datetime(df['install_date'])

df_usa_spend = pd.merge(df[(df['install_date'] >= '2022-04-01') & (df['country_code'] == 'US')], df_spend, on='appsflyer_id')

df_cohort_spend = df_usa_spend.groupby('install_date_x', as_index = False).agg({'user_cost':'sum', 'appsflyer_id':'nunique', 'revenue_usd':'sum', 'ad_revenue_usd':'sum'}).rename(columns = {'install_date_x':'install_date', 'user_cost':'spend_daily', 'appsflyer_id':'installs'})

df_cohort_spend['cpi'] = df_cohort_spend['spend_daily'] / df_cohort_spend['installs']
df_cohort_spend['total_revenue'] = df_cohort_spend['revenue_usd'] + df_cohort_spend['ad_revenue_usd']

df_cohort_spend = df_cohort_spend.fillna(0)

df_cohort_spend['rpi'] = df_cohort_spend['total_revenue'] / df_cohort_spend['installs']
df_cohort_spend['profit'] = df_cohort_spend['total_revenue'] - df_cohort_spend['spend_daily']
df_cohort_spend['roi'] = df_cohort_spend['total_revenue']*100 / df_cohort_spend['spend_daily']

df_cohort_spend
```

Вспомним, что такое LTV и какие есть методы расчета этой метрики ✌

Можно считать, что **накопительный ARPU (cARPU) будет равен LTV** некоторой когорты.
Добавим LTV в наш когортный отчет. И посчитаем LTV помесячно для США.

```python
df_ltv = df_cohort_spend.copy()
df_ltv['install_month'] = df_ltv['install_date'].dt.month

df_ltv = df_ltv.groupby('install_month', as_index = False).agg({'installs':'sum', 'spend_daily':'sum', 'total_revenue':'sum'}).rename(columns = {'spend_daily':'spend_month'})

df_ltv
```

Добавляем LTV и CPI. Рассчитаем помесячную маржинальность через ROI

```python
df_ltv['ltv'] = df_ltv['total_revenue'] / df_ltv['installs']

df_ltv['cpi'] = df_ltv['spend_month'] / df_ltv['installs']

df_ltv['roi'] = df_ltv['total_revenue']*100 / df_ltv['spend_month']

df_ltv
```

# Прогнозирование LTV. Коэффициентная модель 📈

Попробуем рассчитать коэффициентную модель LTV для недельных когорт. При этом рассматривать будем прогноз LTV 14го дня на основе LTV 3го дня жизни когорты.

```python
df_forecast = df[(df['country_code'] == 'US') & (df['install_date'] >= '2022-04-01')].copy()

df_forecast['activity_day'] = (df_forecast['event_date'] - df_forecast['install_date']).dt.days
df_forecast['start_of_week'] = df_forecast['install_date'] - pd.to_timedelta(df_forecast['install_date'].dt.dayofweek, unit='d')

df_forecast
```

Посчитаем для недельных когорт значение LTV на 3ий день жизни когорты.

```python
df_ltv_day3 = df_forecast[df_forecast['activity_day'] <= 3].groupby('start_of_week', as_index = False).agg({'appsflyer_id':'nunique', 'revenue_usd':'sum', 'ad_revenue_usd':'sum'}).rename(columns = {'appsflyer_id':'installs'})
df_ltv_day3['total_revenue'] = df_ltv_day3['revenue_usd'] + df_ltv_day3['ad_revenue_usd']
df_ltv_day3['ltv_day3'] = df_ltv_day3['total_revenue'] / df_ltv_day3['installs']

df_ltv_day3
```

Посчитаем для недельных когорт значение LTV на 14ый день активности.

```python
df_ltv_day14 = df_forecast[df_forecast['activity_day'] <= 14].groupby('start_of_week', as_index = False).agg({'appsflyer_id':'nunique', 'revenue_usd':'sum', 'ad_revenue_usd':'sum'}).rename(columns = {'appsflyer_id':'installs'})
df_ltv_day14['total_revenue'] = df_ltv_day14['revenue_usd'] + df_ltv_day14['ad_revenue_usd']
df_ltv_day14['ltv_day14'] = df_ltv_day14['total_revenue'] / df_ltv_day14['installs']

df_ltv_day14
```

Соединим данные между собой и посчитаем коэффициент LTV 14го дня жизни когорты к 3му дню активности.

```python
df_forecast_ltv = pd.merge(df_ltv_day3[['start_of_week', 'installs', 'ltv_day3']], df_ltv_day14[['start_of_week', 'ltv_day14']], how='left', on='start_of_week')
df_forecast_ltv['ltv_coef'] = df_forecast_ltv['ltv_day14'] / df_forecast_ltv['ltv_day3']

df_forecast_ltv
```

Возьмем медиану по LTV коэффициенту, значение которой будем далее использовать для прогноза. Почему в данном случае следует брать медиану а не среднее?

```python
ltv_coefficient = df_forecast_ltv['ltv_coef'].median()
ltv_coefficient
```

Используем полученный коэффициент для получения прогнозного значения LTV 14го дня жизни когорты по 3му дню активности.

```python
df_forecast_ltv = df_forecast_ltv.rename(columns = {'ltv_day14':'ltv_day14_fact'})
df_forecast_ltv['ltv_day14_forecast'] = df_forecast_ltv['ltv_day3'] * ltv_coefficient

df_forecast_ltv
```

Посмотрим полученные результаты на графике.

```python
import plotly.express as px

fig = px.bar(df_forecast_ltv, x = 'start_of_week', y = ['ltv_day14_fact', 'ltv_day14_forecast'], barmode = 'group')
fig.show()
```
