# Вспомним когортный отчет с второго семинара ⏰

```python
df['event_date'] = pd.to_datetime(df['event_date'])
df['install_date'] = pd.to_datetime(df['install_date'])

df_usa_spend = pd.merge(df[(df['install_date'] >= '2022-04-01') & (df['country_code'] == 'US')], df_spend, on='appsflyer_id')

df_cohort_spend = df_usa_spend.groupby('install_date_x', as_index = False).agg({'user_cost':'sum', 'appsflyer_id':'nunique', 'revenue_usd':'sum', 'ad_revenue_usd':'sum'}).rename(columns = {'install_date_x':'install_date', 'user_cost':'spend_daily', 'appsflyer_id':'installs'})

df_cohort_spend['cpi'] = df_cohort_spend['spend_daily'] / df_cohort_spend['installs']
df_cohort_spend['total_revenue'] = df_cohort_spend['revenue_usd'] + df_cohort_spend['ad_revenue_usd']

df_cohort_spend = df_cohort_spend.fillna(0)

df_cohort_spend['rpi'] = df_cohort_spend['total_revenue'] / df_cohort_spend['installs']
df_cohort_spend['profit'] = df_cohort_spend['total_revenue'] - df_cohort_spend['spend_daily']
df_cohort_spend['roi'] = df_cohort_spend['total_revenue']*100 / df_cohort_spend['spend_daily']

df_cohort_spend
```

При расчете ROI мы использовали полный доход с пользователя - учли его внутренние покупки и рекламную выручку 🤑

# Просмотр рекламы в играх ✨

Что можно узнать из этих данных? 👶

Например в какой стране больше всего рекламной выручки на одного пользователя.

```python
df_ads_daily = df[df['install_date'] >= '2022-04-01'].groupby('country_code', as_index = False).agg({'ad_revenue_usd':'sum', 'appsflyer_id':'nunique'}).rename(columns = {'appsflyer_id':'installs'})

df_ads_daily['ad_revenue_user'] = df_ads_daily['ad_revenue_usd'] / df_ads_daily['installs']

display(df_ads_daily.sort_values(by = 'ad_revenue_usd', ascending = False).head(10))
```

Что можно сказать о просмотре рекламы и рекламной выручке в разных странах в данном проекте?

С чемпионами все понятно, а что с распределением среднего?

```python
df_ads_avg = df[df['install_date'] >= '2022-04-01'].groupby(['appsflyer_id', 'country_code'], as_index = False).agg({'ad_revenue_usd':'sum'})

df_ads_avg = df_ads_avg.groupby('country_code', as_index = False).agg(ad_revenue_mean=pd.NamedAgg(column='ad_revenue_usd', aggfunc='mean'), ad_revenue_median=pd.NamedAgg(column='ad_revenue_usd', aggfunc='median'))

display(df_ads_avg.sort_values(by = 'ad_revenue_mean', ascending = False).head(10))
```

Почему такая разница между средним и медианной? Какие выводы можно сделать из полученных результатов? ⭐

```python
import plotly.express as px

fig = px.bar(df_ads_avg.sort_values(by = 'ad_revenue_mean', ascending = False).head(10), x = 'country_code', y = ['ad_revenue_mean', 'ad_revenue_median'], barmode = 'group')
fig.show()
```

Если бы перед нами стояла задача добавить рекламную монетизацию в когортный отчет непосредственно по данным медиации, то какая основная проблема возникает? ⏹

Все верно, в данных нет атрибуции - поэтому при добавлении их в когортный отчет группировать можно будет только по датам.

Попробуем посчитать одну из основных метрик рекламной монетизации: eCPM.
Напомним:
eCPM = 1000*revenue/impressions

```python
df_ecpm = df_ads.groupby('country_code', as_index = False).agg({'revenue_usd':'sum', 'impressions':'sum'})

df_ecpm['ecpm'] = 1000*df_ecpm['revenue_usd']/df_ecpm['impressions']

display(df_ecpm.sort_values(by = 'ecpm', ascending = False).head(10))
```
